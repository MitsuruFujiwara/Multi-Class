SVMによるクワガタの分類

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import classification_report, accuracy_score
```

データをロード

```python
df = pd.read_csv('dorcus_DL_study.csv')
df_test = pd.read_csv('dorcus_DL_test.csv')
```

```python
df
```

```python
df_test
```

XとYに分割

```python
df_y = df['class']
df_x = df.drop('class', 1)
df_test_y = df_test['class']
df_test_x = df_test.drop('class', 1)
```

SVMで分析

```python
svm = LinearSVC()
result = svm.fit(df_x, df_y)
```

結果を集計

```python
train_pred = svm.predict(df_x)
test_pred = svm.predict(df_test_x)
```

学習用データに対する正解率

```python
accuracy_score(df_y, train_pred)
```

テスト用データに対する正解率

```python
accuracy_score(df_test_y, test_pred)
```

グラフ用にデータを変換

```python
# convert data into vector
def __trY(y):
    for i, t in enumerate(y):
        yield np.eye(1, 3, t-1)

# set data for training
trY = np.array(list(__trY(df_y))).reshape(len(df_y), 3)
trX = np.array(df_x)
```

グラフ用関数を定義

```python
# function for generating decision boundary
def plot_decision_boundary(pred_func):
    # Set min and max values and give it some padding
    x_min, x_max = int(trX[:, 0].min()) - 10, int(trX[:, 0].max()) + 10
    y_min, y_max = int(trX[:, 1].min()) - 10, int(trX[:, 1].max()) + 10
    h = 0.1

    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Predict the function value for the whole gid
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, alpha = 0.4)
    plt.scatter(trX[:,0], trX[:,1], c=trY)
```

結果をプロット

```python
plot_decision_boundary(lambda x: svm.predict(x))
plt.title('decision_boundary_study')
plt.xlabel('days_elapse')
plt.ylabel('weight_gram')
plt.grid()
plt.savefig('decision_boundary_study.png')
plt.show()
```
